```python
stopwords = stopwords.words('english')
stemmer = SnowballStemmer('english')
```

```python
import re
# using regex to remove hyperlinks and other using links
text_clean = "@\S+|https?:\S+|http?:\S|[^A-Za-z0-9]+"
```

```python
# function to cleanup the text
def preprocess(word, stem=False):
  word = re.sub(text_clean, ' ', str(word).lower()).strip()
  tokens = []
  for token in word.split():
    if token not in stopwords:
      if stem:
        tokens.append(stemmer.stem(token))
      else:
        tokens.append(token)
  return " ".join(tokens)
```

```python
df1 = df.copy()
```

```python
df1.head()
```

```python
df1.word = df1.word.apply(lambda x: preprocess(x))
```
